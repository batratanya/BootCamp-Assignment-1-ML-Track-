### 2: Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

```python
df1 = df.query('(location=="Argentina-Catamarca" | location=="Argentina-CABA") & unit=="cases"')
```

```python
df1
```

```python
df1['value'] = pd.to_numeric(df1.value)
```

```python
df2 = df1.groupby('location')
df3 =df2.get_group("Argentina-Catamarca").groupby('report_date').sum()
df4 =df2.get_group("Argentina-CABA").groupby('report_date').sum()
```

```python
caba_final = df4[df4.columns[-1]]
```

```python
caba_final
```

```python
catamarca_final = df3[df3.columns[-1]]
```

```python
catamarca_final
```

```python
report_date=caba_final.index.tolist()
```

```python
plt.style.use('classic')
plt.figure(figsize=(8,8))
plt.plot(report_date,caba_final,label='Argentina-CABA',marker='o')
plt.plot(report_date,catamarca_final,label='Argentina-Catamarca',marker='*')
plt.legend(loc=2,fontsize='medium')
plt.xticks(rotation=60)
```

#### Bar graph for the same comparison: 

```python
barWidth=0.30
r1 = np.arange(len(caba_final))
r2 = [x + barWidth for x in r1]

plt.bar(r1,caba_final,width=barWidth,label='Argentina-CABA',color='b')
plt.bar(r2,catamarca_final,width=barWidth,label='Argentina-Catamarca',color='r')
plt.xticks([r + barWidth for r in range(len(caba_final))],report_date,rotation=90)

plt.legend(loc=2,fontsize='small')
```

### 3. Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

```python
data = df[(df.location=='Argentina-Buenos_Aires') | (df.location=='Argentina-CABA')|(df.location=='Argentina-Catamarca')]
```

```python
data
```

```python
data['value'] = pd.to_numeric(data.value)
```

```python
df0 = data.groupby('location')
```

```python
df0.groups
```

```python
marca=df0.get_group("Argentina-Catamarca").groupby('data_field')['value'].sum().plot(kind='pie',autopct="%1.1f%%")
plt.title('Data Field Value Distribution for Argentina-Catamarca')
plt.legend(loc='best',fontsize='small')
```

```python
caba=df0.get_group("Argentina-CABA").groupby('data_field')['value'].sum().plot(kind='pie',autopct="%1.1f%%")
plt.title('Data Field Value Distribution for Argentina-CABA')
plt.legend(loc='best',fontsize='small')
```

```python
aires=df0.get_group("Argentina-Buenos_Aires").groupby('data_field')['value'].sum().plot(kind='pie',autopct="%1.1f%%")
plt.title('Data Field Value Distribution for Argentina-Buenos_Aires')
plt.legend(loc='lower left',fontsize='small')
```
